```python
dec_df = pd.read_csv("Data/202012.csv")
dec_df.shape
```

```python
def add_date(df, year, month):
    df["Year"] = year;
    df["Month"] = month;
    df.shape;
    return df
```

```python
dec_df_extended = add_date(dec_df,2020,12)
```

```python
dec_df_extended.shape
```

```python
dec_df_extended.head()
```

```python
df = pd.read_csv("Data/202011.csv")
df_extended = add_date(df,2020,11)
df_extended.shape
df_extended.head()
```

```python
df_extended.shape
```

```python
master_df = pd.concat([master_df,dec_df_extended,df_extended])
master_df.head()
```

```python
master_df.tail()
```

```python
df = pd.read_csv("Data/202010.csv")
df_extended = add_date(df,2020,10)
master_df = pd.concat([master_df,df_extended])
master_df.shape
```

```python
for i in range(3,10):
    url = "Data/20200" + str(i) + ".csv";
    print(url);
    df = pd.read_csv(url)
    df_extended = add_date(df,2020,i)
    master_df = pd.concat([master_df,df_extended])
    print(master_df.shape)
    i+=1
```

```python
master_df.drop_duplicates()
```

```python
master_df.to_csv("Data/CoronaYear.csv")
```

# Create data set for year prior to Corona Virus (March 2019 - Feb 2020)

```python
url = ""
df = []
df_extended = []
master_df = []

for i in range(3,13):
    url = "Data/2019" + str(i) + ".csv";
    print(url);
    df = pd.read_csv(url)
    df_extended = add_date(df,2019,i)
    if i > 3:
        master_df = pd.concat([master_df,df_extended])
    elif i == 3:
        master_df = df_extended
    print(master_df.shape)
    i+=1

i = 1
for i in range(1,3):
    url = "Data/20200" + str(i) + ".csv";
    print(url);
    df = pd.read_csv(url)
    df_extended = add_date(df,2020,i)
    master_df = pd.concat([master_df,df_extended])
    print(master_df.shape)
    i+=1
```

```python
master_df.to_csv("Data/PreCoronaYear.csv")
```
